````python
cnts['clip_enrichment'] = cnts['CLIP-35L33G.bam'] / cnts['RNA-control.bam']
cnts['rden_change'] = (cnts['RPF-siLin28a.bam'] / cnts['RNA-siLin28a.bam']) / (cnts['RPF-siLuc.bam'] / cnts['RNA-siLuc.bam'])
cnts.head()
````

````python
# ì—¬ê¸°ì— ì½”ë“œë¥¼ ìƒˆë¡œ ì‘ì„±í•´ì„œ ë…¼ë¬¸ì˜ ê·¸ë¦¼ê³¼ ë¹„ìŠ·í•˜ê²Œ ë§Œë“¤ì–´ ë´…ë‹ˆë‹¤.

# í•„ìš”í•œ Columns ëŠ” ìœ„ì—ì„œ ë§Œë“¤ì—ˆìŒ
# cnts['clip_enrichment'] = cnts['CLIP-35L33G.bam'] / cnts['RNA-control.bam']
# cnts['rden_change'] = (cnts['RPF-siLin28a.bam'] / cnts['RNA-siLin28a.bam']) / (cnts['RPF-siLuc.bam'] / cnts['RNA-siLuc.bam'])

import numpy as np
from scipy.stats import pearsonr
import matplotlib.pyplot as plt

# ------------------------------------------------------------------------------
# Filter out low-count genes
#
# Genes with very low read counts are unreliable, so we exclude them.
# Adjust threshold as needed
# ------------------------------------------------------------------------------

threshold = 10  # or 20 for stricter filter

cnts_filtered = cnts[
    (cnts['CLIP-35L33G.bam'] > threshold) &
    (cnts['RNA-control.bam'] > threshold) &
    (cnts['RPF-siLin28a.bam'] > threshold) &
    (cnts['RNA-siLin28a.bam'] > threshold) &
    (cnts['RPF-siLuc.bam'] > threshold) &
    (cnts['RNA-siLuc.bam'] > threshold)
].copy()

# ------------------------------------------------------------------------------
# Compute CLIP enrichment and ribosome density change
#
# CLIP enrichment = CLIP reads / total RNA reads
# Ribosome density change = (RPF/RNA) ratio under Lin28a knockdown vs control
# ------------------------------------------------------------------------------
cnts_filtered['clip_enrichment'] = cnts_filtered['CLIP-35L33G.bam'] / cnts_filtered['RNA-control.bam']
cnts_filtered['rden_change'] = (
    (cnts_filtered['RPF-siLin28a.bam'] / cnts_filtered['RNA-siLin28a.bam']) /
    (cnts_filtered['RPF-siLuc.bam'] / cnts_filtered['RNA-siLuc.bam'])
)
# ------------------------------------------------------------------------------
# Remove invalid values and log-transform
#
# Drop infinite/NaN values, and exclude any zeros or negatives because
# log2(0) is undefined
# ------------------------------------------------------------------------------
cnts_filtered = cnts_filtered.replace([np.inf, -np.inf], np.nan).dropna(subset=['clip_enrichment', 'rden_change'])
cnts_filtered = cnts_filtered[(cnts_filtered['clip_enrichment'] > 0) & (cnts_filtered['rden_change'] > 0)]

cnts_filtered['log2_clip'] = np.log2(cnts_filtered['clip_enrichment'])
cnts_filtered['log2_rden'] = np.log2(cnts_filtered['rden_change'])

# ------------------------------------------------------------------------------
# Calculate Pearson correlation coefficient
#
# This quantifies the linear relationship between CLIP enrichment and ribosome
# density change across all genes
# ------------------------------------------------------------------------------
r_val, _ = pearsonr(cnts_filtered['log2_clip'], cnts_filtered['log2_rden'])

# ------------------------------------------------------------------------------
# Plot the scatterplot similar to Figure 4D
#
# Each point represents a gene. The x-axis is CLIP enrichment (log2),and the
# y-axis is ribosome density change (log2)
# ------------------------------------------------------------------------------
plt.figure(figsize=(5, 5))
plt.scatter(cnts_filtered['log2_clip'], cnts_filtered['log2_rden'],
            color='black', s=1, alpha=0.5)

plt.xlabel('LIN28A CLIP enrichment (logâ‚‚)', fontsize=10)
plt.ylabel('Ribosome density change\nupon Lin28a knockdown (logâ‚‚)', fontsize=10)
plt.title('D  CLIP and ribosome footprinting upon\nLin28a knockdown', loc='left', fontsize=12, weight='bold')

plt.text(3, -1.8, f"$r$ = {r_val:.4f}", fontsize=10)

plt.xlim(-6, 6)
plt.ylim(-3, 3)
plt.xticks(np.arange(-6, 5, 2))
plt.yticks(np.arange(-2, 3, 1))
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.4)
plt.tight_layout()
plt.show()
````
